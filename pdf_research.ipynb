```python
from typing import Union
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
path = 'data/'

df = pd.read_csv(path+'patient_data.csv', index_col = 0)

df.birth_date = pd.to_datetime(df.birth_date, format = "%d.%m.%Y")

now = datetime.now()
df['age'] = df.birth_date.apply(lambda x: relativedelta(now, x).years)
df['birth_year'] = df.birth_date.apply(lambda x: x.year)
```
