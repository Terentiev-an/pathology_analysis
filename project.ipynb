```python
import pandas as pd
import numpy as np

from natasha import Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger, NewsSyntaxParser, NewsNERTagger,NamesExtractor, Doc
segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)
```

```python
lems = []
for item in df['microscop_desc'].to_list():
  doc = Doc(item)
  # Инициализация парсинга?
  doc.segment(segmenter) 
  #print(doc.segment)
  
  # Разбивка на предложения
  doc.sents
  #print(doc.sents)   
  
  # Разбивка на Токены(слова)
  doc.tokens
  #print(doc.tokens)  

  # Lematization , тут переводим слова в нормальный вид, и загоняем в список
  lem = []
  for tok in doc.tokens:
    doc.tag_morph(morph_tagger)
    tok.lemmatize(morph_vocab)
    lem.append(tok.lemma)
  lems.append(lem)
```

```python
df['temp'] = lems
```

```python






segmenter = Segmenter()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)

for item in df['microscop_desc'].to_list():
  doc = Doc(item)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)
  doc.parse_syntax(syntax_parser)

  sent = doc.sents[0]
  sent.morph.print()
  print('#################################################################')
  #print()
  #print(doc.tokens)
  #print()
  #print('#################################################################')
  print()
  sent.syntax.print()
  print()
  print('#################################################################')
  print()
  doc.tag_ner(NewsNERTagger(emb))
  doc.ner.print()
  print()
  print('#################################################################')
```
